What is the average rating for each product category

```python
# Convert rating column to numeric (if not already)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Average rating per category
avg_rating = df.groupby('category')['rating'].mean().reset_index()

print(avg_rating)
```

```python
# Convert rating_count to numeric (remove commas if needed)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',','').astype(float)

# Sort by rating_count (highest first)
df_sorted = df.sort_values(by='rating_count', ascending=False)

# For each category, get the top product(s) by rating_count
top_products = df_sorted.groupby('category').head(1)   # top 1 per category

# If you want top 3 products per category
top3_products = df_sorted.groupby('category').head(3)

print("Top product by rating_count in each category:\n", top_products[['category','product_name','rating_count']])
print("\nTop 3 products by rating_count in each category:\n", top3_products[['category','product_name','rating_count']])
```

4. How does the average discount percentage vary across categories?


```python
print(df['discount_percentage'])
```

```python
# Remove % if present and convert to float
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "", regex=True)

# Convert to float
df["discount_percentage"] = df["discount_percentage"].astype(float)

# (Optional) Convert into real percentage (0.25 instead of 25)
df["discount_percentage"] = df["discount_percentage"] / 100

```

```python
print(df['discount_percentage'])
```

```python
# Group by category and calculate average discount %
category_discounts = df.groupby("category")["discount_percentage"].mean().reset_index()
```

```python
# Sort for better visualization
category_discounts = category_discounts.sort_values(by="discount_percentage", ascending=False)
```

```python
# Display table
print(category_discounts)
```

```python
# Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(category_discounts["category"], category_discounts["discount_percentage"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("Average Discount Percentage")
plt.title("Average Discount % by Category")
plt.show()
```

6. What are the most popular product keywords?


```python
import re
from collections import Counter
# Combine all product names into a single string
all_names = " ".join(df["product_name"].astype(str))

# Convert to lowercase and remove punctuation/numbers
all_names = re.sub(r"[^a-zA-Z\s]", "", all_names.lower())

# Split into words
words = all_names.split()

# Optionally remove common stopwords
stopwords = {"and", "with", "for", "the", "of", "in", "a"}  # you can expand this list
keywords = [word for word in words if word not in stopwords]

# Count frequency
keyword_counts = Counter(keywords)

# Top 20 most common keywords
top_keywords = keyword_counts.most_common(20)
print(top_keywords)
```

```python

# Optional: visualize
words, counts = zip(*top_keywords)
plt.figure(figsize=(12,6))
plt.bar(words, counts)
plt.xticks(rotation=45, ha='right')
plt.title("Top Product Keywords")
plt.show()
```

8. What is the correlation between discounted_price and rating?

```python
# Ensure the columns are numeric
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors='coerce')
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')

# Drop rows with missing values in these columns
df_clean = df.dropna(subset=["discounted_price", "rating"])

# Calculate correlation
correlation = df_clean["discounted_price"].corr(df_clean["rating"])
print("Correlation between discounted_price and rating:", correlation)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_clean["discounted_price"], df_clean["rating"], alpha=0.5)
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.title("Discounted Price vs Rating")
plt.show()
```

9. What are the Top 5 categories based on the highest ratings?

```python
# Ensure rating column is numeric
df["rating"] = pd.to_numeric(df["rating"], errors='coerce')

# Drop rows with missing ratings
df_clean = df.dropna(subset=["rating"])

# Group by category and calculate average rating
category_ratings = df_clean.groupby("category")["rating"].mean().reset_index()

# Sort descending by rating
category_ratings = category_ratings.sort_values(by="rating", ascending=False)

# Top 5 categories
top_5_categories = category_ratings.head(5)
print(top_5_categories)
```
